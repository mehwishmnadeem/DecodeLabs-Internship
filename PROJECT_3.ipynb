```python
movies = {
    "Action": [
        "John Wick",
        "Mission Impossible",
        "Mad Max: Fury Road",
        "The Dark Knight"
    ],

    "Comedy": [
        "The Hangover",
        "Free Guy",
        "Jumanji",
        "Rush Hour"
    ],

    "Romance": [
        "Titanic",
        "The Notebook",
        "La La Land",
        "Me Before You"
    ],

    "Horror": [
        "The Conjuring",
        "Insidious",
        "IT",
        "A Quiet Place"
    ],

    "Science Fiction": [
        "Interstellar",
        "Inception",
        "The Matrix",
        "Avatar"
    ],

    "Animation": [
        "Toy Story",
        "Frozen",
        "Finding Nemo",
        "Kung Fu Panda"
    ]
}
def display_genres():
    print("\nAvailable Genres:")
    print("-" * 30)

    for genre in movies:
        print("•", genre)

    print("-" * 30)
def recommend_movies(user_choice):

    found = False

    for genre, movie_list in movies.items():

        if user_choice.lower() == genre.lower():

            found = True

            print("\nRecommended Movies")
            print("=" * 40)

            for i, movie in enumerate(movie_list, start=1):
                print(f"{i}. {movie}")

            print("=" * 40)

            print(f"\nTotal Recommendations: {len(movie_list)}")

    if not found:
        print("\nSorry!")
        print("No recommendations found for that genre.")

def main():

    print("=" * 55)
    print("      SMART MOVIE RECOMMENDATION SYSTEM")
    print("=" * 55)

    name = input("Enter your name: ")

    print(f"\nWelcome, {name}!")

    display_genres()

    choice = input("\nEnter your preferred genre: ")

    recommend_movies(choice)

    print("\nThank you for using the Recommendation System!")
    print("Have a great day!")

main()
```
